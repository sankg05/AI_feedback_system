````python
import pandas as pd
import os
import numpy as np
import json
import re
import google.generativeai as genai
from tqdm import tqdm
from sklearn.metrics import accuracy_score, mean_absolute_error
import requests
````

````python
ZERO_SHOT_PROMPT_TEMPLATE = """You are a sentiment classification assistant.

Read the Yelp review below and assign a star rating from 1 to 5 based on
the overall sentiment and experience described.

Return your response strictly in valid JSON with the following format:
{{
  "predicted_stars": integer,
  "explanation": string
}}

Rules:
- predicted_stars must be an integer between 1 and 5
- explanation should be brief (1–2 sentences)
- Do not include any text outside the JSON

Review:
"{review_text}"
"""
````

````python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

MODEL_NAME = "mistralai/Mistral-7B-Instruct-v0.3"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto",
    load_in_4bit=True
)

model.eval()
````

````python
def safe_parse_model_output(text):
    import json, re

    text = text.replace("```json", "").replace("```", "").strip()

    # 🔥 FIND ALL JSON OBJECTS
    matches = re.findall(r"\{[\s\S]*?\}", text)
    if not matches:
        return None

    # ✅ USE THE LAST ONE (MODEL ANSWER)
    raw_json = matches[-1]

    try:
        data = json.loads(raw_json)
    except Exception:
        return None

    # Flexible schema
    stars = data.get("stars") or data.get("predicted_stars")
    explanation = data.get("explanation", "")

    if stars is None:
        return None

    try:
        stars = int(stars)
    except Exception:
        return None

    if not 1 <= stars <= 5:
        return None

    return {
        "predicted_stars": stars,
        "explanation": explanation,
        "raw_json": raw_json
    }
````

# COT MODEL

````python
def classify_review_zero_shot(review_text: str):
    prompt = ZERO_SHOT_PROMPT_TEMPLATE.format(review_text=review_text)

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=150,
            temperature=0.2,
            do_sample=False,
            eos_token_id=tokenizer.eos_token_id
        )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # ✅ SAFE PARSE HERE
    parsed = safe_parse_model_output(decoded)

    if parsed is None:
        print("⚠️ Parsing failed")
        return None

    return parsed

````

````python
df = pd.read_csv("/content/yelp.csv")
sampled_df = df.sample(n=200, random_state=42)
sampled_df.head(2)
````

````python
results = []
failed = 0

for _, row in tqdm(sampled_df.iterrows(), total=200):
    prediction1 = classify_review_zero_shot(row["text"])
    if prediction1 is None:
        failed += 1
        results.append({
            "text": row["text"],
            "true_stars": row["stars"],
            "predicted_stars": None,
            "explanation": None,
            "raw_model_json": None
        })
        continue

    results.append({
        "text": row["text"],
        "true_stars": row["stars"],
        "predicted_stars": prediction1["predicted_stars"],
        "explanation": prediction1["explanation"],
        "raw_model_json": prediction1["raw_json"]
    })
````

# ACCURACY

````python
import pandas as pd

results_df = pd.DataFrame(results)
results_df.to_csv("zero_shot_results.csv", index=False)

valid = results_df.dropna()

accuracy = (valid["true_stars"] == valid["predicted_stars"]).mean()
mae = (valid["true_stars"] - valid["predicted_stars"]).abs().mean()

print(f"Total samples: {len(results_df)}")
print(f"Valid predictions: {len(valid)}")
print(f"Failed predictions: {failed}")
print(f"Accuracy: {accuracy:.3f}")
print(f"MAE: {mae:.3f}")
````
